```python
def movie_profile_calculator(data, reference):
    vec = [0]*len(reference)
    for i in range(len(data)):
        if data[i] in reference:
            idx = reference.index(data[i])
            vec[idx]=1
    return np.array(vec)
```

```python
def pearson_corr(x, y):
    
    result = np.array([np.array([x[i], y[i]]) for i in range(len(x)) if x[i] != 0 and y[i] != 0])
    if len(result) == 0:
        return 0
    r_u, r_v = result.T[0], result.T[1]
    if np.var(r_u) == 0 or np.var(r_v) == 0:
        corr = -math.inf
    else:
        corr = stats.pearsonr(r_u, r_v)[0]
    return corr

def cosine_corr(x, y):
    
    result = np.array([np.array([x[i], y[i]]) for i in range(len(x)) if x[i] != 0 and y[i] != 0])
    if len(result) == 0:
        return 0
    r_u, r_v = result.T[0], result.T[1]
    # Calculate numerator of Pearson correlation coefficient
    numerator = np.dot(r_u, r_v)

    # Calculate denominator of Pearson correlation coefficient
    denominator = np.sqrt(np.sum(np.square(r_u))) * np.sqrt(np.sum(np.square(r_v)))

    # Calculate Pearson correlation coefficient
    corr = numerator / denominator
    return corr


def Modified_Jaccard(x, y):
    #common = len(rating_matrix_df.columns[(rating_matrix_df.loc[user1].gt(0) & rating_matrix_df.loc[user2].gt(0))])
    common = np.sum(np.array([1 for i in range(len(x)) if x[i] != 0 and y[i] != 0]))
    I_u = np.count_nonzero(x)
    I_v = np.count_nonzero(y) + 1
    
    return common/ (I_u * I_v)

def Proximity(a, b):
    expo = math.exp(-abs(a-b))
    return expo/(1 + expo)

def Significance(a, b, median):
    expo = math.exp(-abs(a-median)*abs(b-median))
    return 1/(1 + expo)

def Singularity(a, b, mean):
    expo = math.exp(-abs(((a+b)/2) - mean))
    return expo/(1 + expo)

def PSSSim(x, y, mean, median=3):
    result = np.array([np.array([x[i], y[i]]) for i in range(len(x)) if x[i] != 0 and y[i] != 0])
    if len(result) == 0:
        return 0
    r_u, r_v = result.T[0], result.T[1]
    Proximity_res = [Proximity(x, y) for x, y in zip(r_u, r_v)]
    Significance_res = [Significance(x, y, median) for x, y in zip(r_u, r_v)]
    Singularity_res = [Singularity(x, y, mean) for x, y in zip(r_u, r_v)]
    
    return np.prod(np.array(Proximity_res) + np.array(Significance_res) + np.array(Singularity_res))




def URPSim(x, y):
    result = np.array([np.array([x[i], y[i]]) for i in range(len(x)) if x[i] != 0 and y[i] != 0])
    if len(result) == 0:
        return 0
    #r_u, r_v = result.T[0], result.T[1]
    r_u, r_v = x[x>0], y[y>0]
    expo = math.exp(-abs(np.mean(r_u) - np.mean(r_v)) * abs(np.std(r_u) - np.std(r_v)))
    
    return expo/(1 + expo)

def plot_fig(data_df):
    sns.lineplot(data= data_df[["R2_score", "RMSE", "MAE"]])
    plt.xlabel("K")
    plt.ylabel("Performance Metric")
    plt.show()
```

```python
def similarity_computation(target_user, target_movie, rating_matrix_df):

    data_filtered = rating_matrix_df.loc[rating_matrix_df[target_movie]!=0]
    target_movie_mean_rating = data_filtered[target_movie].mean() - data_filtered.loc[target_user][target_movie]
    target_user_data = data_filtered.loc[target_user].to_numpy()
    target_user_mean_rating = np.mean(target_user_data[target_user_data>0])
    data_filtered = data_filtered.drop([target_movie], axis = 1)

    common_users = data_filtered.index

    pearson_corr_sim = {}
    cosine_corr_sim = {}
    Modified_Jaccard_sim = {}
    PSS_Sim = {}
    URP_Sim = {}
    for pair in itertools.combinations(common_users, 2):
        #print(pair)
        if pair[0] == target_user or pair[1] == target_user:
            if pair[0] == target_user:
                x = data_filtered.loc[pair[0]].to_numpy()
                y = data_filtered.loc[pair[1]].to_numpy()
                other_user = pair[1]
            else:
                x = data_filtered.loc[pair[1]].to_numpy()
                y = data_filtered.loc[pair[0]].to_numpy()
                other_user = pair[0]
            pearson_corr_sim[other_user] = pearson_corr(x, y)
            cosine_corr_sim[other_user] = cosine_corr(x, y)
            Modified_Jaccard_sim[other_user] = Modified_Jaccard(x, y)
            PSS_Sim[other_user] = PSSSim(x, y,  mean = target_movie_mean_rating, median=3)
            URP_Sim[other_user] = URPSim(x, y)

    data = {key: [pearson_corr_sim[key], cosine_corr_sim[key], Modified_Jaccard_sim[key], PSS_Sim[key], URP_Sim[key]] for key in pearson_corr_sim}

    data_df = pd.DataFrame.from_dict(data = data, 
                                  orient='index',
                                  columns=['pearson', 'cosine', 'Modified Jaccard', 'PSS', 'URP'])
    data_df["JPSS"] = data_df["PSS"] * data_df["Modified Jaccard"]
    data_df["NHSM"] = data_df["JPSS"] * data_df["URP"]
    
    #data_df.sort_values(by=['pearson'], ascending=False, inplace=True)
    return data_df, target_user_mean_rating
```

```python
def train_test(corr_df,similarity_considered, target_user,target_user_mean_rating, target_movie,rating_matrix_df, k=5):
    
    k = min(k, corr_df.shape[0])
    corr_df.sort_values(by=[similarity_considered], ascending=False, inplace=True)
    #predict_rating = corr_df["pearson"].head(k)
    top_k_users = corr_df.head(k).index
    top_k_data = rating_matrix_df.loc[top_k_users].to_numpy()
    
    rating_given = rating_matrix_df.loc[top_k_users][target_movie].to_numpy()
    #print(rating_given)
    top_k_user_mean = np.where(top_k_data > 0, top_k_data, 0).sum(axis=1)/np.count_nonzero(top_k_data, axis = 1)
    #print(top_k_user_mean)
    
    
    adj_rating_given = rating_given - top_k_user_mean
    #print(adj_rating_given)
    similarity = corr_df[similarity_considered].head(k).to_numpy()
    #print(similarity)
    if similarity_considered != 'cosine':
        if similarity.size != 0 and adj_rating_given.size !=0:
            try:
                if np.sum(np.absolute(similarity))!=0:
                    val = np.dot(adj_rating_given, similarity)/np.sum(np.absolute(similarity))
                else:
                    val = 0
                prediction = target_user_mean_rating + val
            except Exception as e:
                prediction = -1
                print(similarity)
                print(adj_rating_given)
                print(e)
        else:
            prediction = -1
    
    if similarity_considered == 'cosine':
        
        if similarity.size != 0 and rating_given.size !=0:
            try:
                if np.sum(np.absolute(similarity))!=0:
                    val = np.dot(rating_given, similarity)/np.sum(np.absolute(similarity))
                else:
                    val = 0
                prediction = val
            except Exception as e:
                prediction = -1
                print(similarity)
                print(adj_rating_given)
                print(e)
        else:
            prediction = -1
        
    if math.isnan(prediction):
        prediction = -1
    return prediction
```

```python
def validation(X_test, similarity_considered, target_user, rating_matrix_df):
    r2_score_list = []
    mae_list = []
    rmse_list = []
    K=[]
    pred_list = []
    test_list = []

    for i in range(2,X_test.shape[0]-1):
        y_pred = []
        
        for movie in X_test:
            target_movie = movie[0]
            data_df, target_user_mean_rating = similarity_computation(target_user, target_movie, rating_matrix_df)
            '''
            if similarity_considered == 'novel':
                pred = (train_test(data_df, target_user,target_user_mean_rating, target_movie, rating_matrix_df, k= i, similarity_considered = 'pearson') + 
                train_test(data_df, target_user,target_user_mean_rating, target_movie, rating_matrix_df, k= i, similarity_considered = 'NHSM'))/2
                
                y_pred.append(pred)
            else:
            '''
            pred = train_test(data_df,similarity_considered, target_user,target_user_mean_rating, target_movie, rating_matrix_df, k= i)
            y_pred.append(pred)
        #print(len(y_pred), y_pred)
        #print(y_test.shape)
        res = pd.DataFrame({ "Actual": y_test,
                            "Predicted": y_pred
        })
        res = res.loc[res["Predicted"]!=-1]

        y_test_modified = res["Actual"]
        y_pred = res["Predicted"]
        
        pred_list.append(y_pred.to_numpy())
        test_list.append(y_test_modified.to_numpy())

        r2_score_list.append(r2_score(y_test_modified, y_pred))
        mae_list.append(mean_absolute_error(y_test_modified,y_pred))
        rmse_list.append(mean_squared_error(y_test_modified, y_pred, squared = False))
        K.append(i)
        #print("K", i)
        #print("R^2 : ", r2_score_list[-1])
        #print("MAE :", mae_list[-1])
        #print("RMSE:",rmse_list[-1])

    result1 = pd.DataFrame({"R2_score": r2_score_list,
                           "MAE" : mae_list,
                           "RMSE" : rmse_list
                          }, index = K)
    
    prediction_df = pd.DataFrame({"Predicted Values": pred_list,
                                  "Test Values": test_list
                          }, index = K)
    print(similarity_considered,'\n')
    print(result1,'\n')
    
    plot_fig(result1)
    
    return prediction_df
```
