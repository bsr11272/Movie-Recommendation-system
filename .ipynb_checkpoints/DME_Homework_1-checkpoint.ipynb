```python
for col in df:
    print(str(col)+" :", df[col].unique())
```

```python
thresh = 10
na_greater_thresh_features = [ col for col in df.columns if df[col].isnull().sum() > thresh ]
na_greater_thresh_features
```

```python
for feature in na_greater_thresh_features:
  df[feature] = df[feature].fillna(df.groupby('Class')[feature].transform('mean'))
df
```

```python
100*(df.isna().sum()/df.shape[0])
```

```python
thresh = 10
na_less_thresh_features = [ col for col in df.columns if df[col].isnull().sum() < thresh ]
na_less_thresh_features
```

```python
for feature in na_less_thresh_features:
  df[feature] = df.groupby(['Class'])[feature].ffill()
df
```

**TODO1:** Is LOR score related to CGPA? Use visualization to back up your answer.

```python
df["LOR "] = (df["LOR "] - df["LOR "].mean())/df["LOR "].std()
df["CGPA"] = (df["CGPA"] - df["CGPA"].mean())/df["CGPA"].std()
df.plot.scatter(x='LOR ', y='CGPA')
```
