## 3 types of averages

### 1- simple average but fails in considering bias of one movie over other
### 2- Jaccard similarity 
### 3- weighted average

```python
def cosine_similarity_man(profile_vec, reference):
    #reference[np.isnan(reference)] = 0
    return reference.dot(profile_vec)/(np.linalg.norm(reference)*np.linalg.norm(profile_vec))
```

```python
def recommendation(user, user_profile_df, movies_df):
    preference = user_profile_df.iloc[user-1][1]
    data_movie_profile = pd.DataFrame(movies_df["movieId"])
    data_movie_profile["Movie Profile Vector"] = movies_df["Movie Profile Vector"]
    data_movie_profile["Sim_index"] = data_movie_profile["Movie Profile Vector"].apply(cosine_similarity_man, reference = preference)
    return data_movie_profile.sort_values("Sim_index",ascending=False)
```

```python
def predict(ratings_df,rating_matrix_df, user_similarity, X_test,y_test, user, k=5):
    
    movies_to_rate = X_test
    value= []
    for movie in movies_to_rate:
        users_rated_movie = ratings_df[ratings_df["movieId"] == movie][["userId"]]["userId"].tolist()
        idx = [x - 1 for x in users_rated_movie]
        neighbours_similarity = pd.DataFrame({ "Neighbor": users_rated_movie,
                                          "Sim_index": user_similarity[user-1][idx]})
        neighbours_similarity.sort_values(by=['Sim_index'], inplace=True, ascending=False)
        
        neighbours_similarity = neighbours_similarity[neighbours_similarity["Neighbor"]!=user]
        #print(neighbours_similarity)
        k = min(k, neighbours_similarity.shape[0])

        rating_given = rating_matrix_df.iloc[neighbours_similarity["Neighbor"]-1][movie].head(k)
        #print(rating_given)
        neighbours_similarity = neighbours_similarity["Sim_index"].head(k)
        #print(neighbours_similarity)
        if sum(neighbours_similarity)!=0:
            value.append(np.dot(np.array(neighbours_similarity.tolist()),np.array(rating_given.tolist()))/sum(neighbours_similarity.tolist()))
        #print((sum(neighbours_similarity.tolist())))
        else:
            value.append(None)
    rating_prediction = pd.DataFrame({"MovieId": X_test,
                                      "Predicted_rating": value,
                                      "Actual Y": y_test
    })
    return rating_prediction
```

```python
user =414

user_similarity = cosine_similarity(rating_matrix_df)
data_user = ratings_df[ratings_df["userId"]== user][["movieId","rating"]]
X_train, X_test, y_train, y_test = train_test_split(data_user["movieId"],data_user["rating"] , test_size=0.1, random_state=42)

r2_score_list = []
mae_list = []
rmse_list = []
K=[]
for i in range(2,100):
    df = predict(ratings_df,rating_matrix_df, user_similarity, X_test, y_test, user, k = i)
    #print(df)
    y_test = df["Actual Y"].fillna(0)
    y_pred = df["Predicted_rating"].fillna(0)
    r2_score_list.append(r2_score(y_test, y_pred))
    mae_list.append(mean_absolute_error(y_test,y_pred))
    rmse_list.append(mean_squared_error(y_test, y_pred, squared = False))
    K.append(i)
    #print("K", k)  
    #print("R^2 : ", r2_score_list[-1])
    #print("MAE :", mae_list[-1])
    #print("RMSE:",rmse_list[-1])

result = pd.DataFrame({"R2_score": r2_score_list,
                       "MAE" : mae_list,
                       "RMSE" : rmse_list
                      }, index = K)
result
```
